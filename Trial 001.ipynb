```python
plt.style.use('seaborn-whitegrid')
plt.rcParams['figure.figsize'] = [20,10]
```

```python
#*-*-*-*-*-*-*-*-*-*-
# Creating a scatter plot between BA and W
#*-*-*-*-*-*-*-*-*-*-

plt.scatter(df['BA'], df['W'], s = 60) # 's' for 's'ize of data points.
plt.ylabel('Number of Wins', fontsize = 14)
plt.xlabel('Batting Average', fontsize = 14)
plt.title('Number of Wins vs Batting Average', fontsize = 20)
```

```python
import seaborn as sns

sns.regplot(df['BA'], df['W'], ci = None)
plt.ylabel('Number of Wins', fontsize = 14)
plt.xlabel('Batting Average', fontsize = 14)
plt.title('Number of Wins vs Batting Average', fontsize = 20)
```

```python
from scipy.stats.stats import pearsonr # Pearson’s
from scipy.stats.stats import spearmanr # Spearmans

r, p = pearsonr(df['BA'], df['W'])
r, p
print("r=", round(r,3))
```

```python
print("r =", round(r, 3))

if p < .001:
  print("p <.001")
elif p< .01:
    print("p <.01")
elif p< .05:
    print("p <.05")
else:
    print("Not significant")
```
