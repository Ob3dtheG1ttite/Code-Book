```python
from scipy.stats.stats import pearsonr # Pearson’s 
from scipy.stats.stats import spearmanr # Spearman’s 
```

```python
r, p = pearsonr(df['af'], df['tmin'])
print("r =", round(r, 3)) 
 
if p < .001: 
  print("p <.001") 
elif p <.01: 
  print("p <.01") 
elif p <.05: 
  print("p <.05") 
else: 
  print("Not significant")
```

```python
import statsmodels.api as sm
y = df['rain']     # Assigning df.rain to y, the DV 
x = df['tmax']    # Assigning df.tmax to x, the IV 
 
x = sm.add_constant(x)   # Assigning the intercept to x, the IV 
mdl = sm.OLS(y, x).fit() # Build and fit the OLS model (‘mdl’) 
 
print(mdl.summary())  
```
