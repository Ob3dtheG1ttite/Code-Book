```python
# df.drop(['Hours','streams','Residence_ISO3' , 'Reference' , 'League' , 'GAD_T' , 'SWL_T' , 'SPIN_T'],axis=1,inplace=True) Calculating clusters
df3 = df1[['GAD_T'  , 'SWL_T' , 'SPIN_T' ]]
df4 = df1[[ 'Age' ,  'Hours' , 'streams' ]]
pc1 = PCA(n_components=2)
pc2 = PCA(n_components=2)
x1 = pc1.fit_transform(df3)
x2 = pc2.fit_transform(df4)
x = x1 + x2
```

```python
#determine number of clusters
WCSS = []
for i in range(1,12):
    model = KMeans(n_clusters = i,init = 'k-means++')
    model.fit(x)
    WCSS.append(model.inertia_) #inertia --> error
fig = plt.figure(figsize = (7,7))
plt.plot(range(1,12),WCSS, linewidth=4, markersize=12,marker='o',color = 'red')
plt.xticks(np.arange(12))
plt.xlabel("Number of clusters")
plt.ylabel("WCSS")
plt.show()
```

```python
#Clustering into five groups
model = KMeans(n_clusters = 5, init = "k-means++", max_iter = 300, n_init = 40, random_state = 0)
y_clusters = model.fit_predict(x)
pd.Series(y_clusters).value_counts().plot(kind='bar');
```

```python
#Cluster Group graph
plt.scatter(x[y_clusters == 0, 0], x[y_clusters == 0, 1], s = 60, c = 'blue', label = 'Cluster1')
plt.scatter(x[y_clusters == 1, 0], x[y_clusters == 1, 1], s = 60, c = 'orange', label = 'Cluster2')
plt.scatter(x[y_clusters == 2, 0], x[y_clusters == 2, 1], s = 60, c = 'indigo', label = 'Cluster3')
plt.scatter(x[y_clusters == 3, 0], x[y_clusters == 3, 1], s = 60, c = 'violet', label = 'Cluster4')
plt.scatter(x[y_clusters == 4, 0], x[y_clusters == 4, 1], s = 60, c = 'green', label = 'Cluster5') 

plt.legend()
plt.show()
```
